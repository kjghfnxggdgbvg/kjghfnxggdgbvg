Q20) Calculate probability from the given dataset for the below cases

P(MPG>38)

P(MPG<40)

P (20<MPG<50)

```python
car = pd.read_csv('Cars.csv')
car.head()
```

```python
car.tail()
```

```python
car.info()
```

```python
car.describe()
```

```python
mean_mpg = car.MPG.mean()
std_mpg = car.MPG.std()
print('Mean :',mean_mpg,'\nStandard Deviation :',std_mpg)
```

```python
prob_mpg_greater_than_38 = np.round(1 - stats.norm.cdf(38,loc=mean_mpg,scale=std_mpg),2)
print('P(MPG>38) :',prob_mpg_greater_than_38)
```

```python
prob_mpg_less_than_40 = np.round(stats.norm.cdf(40,loc=mean_mpg,scale=std_mpg),2)
print('P(MPG<40) :',prob_mpg_less_than_40)
```

```python
prob_mpg_greater_than_20 = np.round(1 - stats.norm.cdf(20,loc=mean_mpg,scale=std_mpg),2)
prob_mpg_less_than_50 = np.round(stats.norm.cdf(50,loc=mean_mpg,scale=std_mpg),2)
prob_mpg_between_20_and_50 = np.round(prob_mpg_less_than_50 - prob_mpg_greater_than_20,2)
print('P(20<MPG<50) :',prob_mpg_between_20_and_50)
```
