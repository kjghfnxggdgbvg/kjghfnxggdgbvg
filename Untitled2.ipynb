```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
book = pd.read_csv('book.csv')
book.head()
```

```python
book.shape
```

APIRORI ALGORITHM

```python
frequent_itemsets = apriori(book, min_support = 0.1, use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets.sort_values("support", ascending = False).head()
```

create the association rules from the frequent itemset

```python
rules = association_rules(frequent_itemsets, metric= "confidence", min_threshold=0.4)
print(rules.head())
```

```python
rules[(rules.support > 0.015) & (rules.confidence > 0.4)].sort_values("confidence", ascending = False).shape
```

```python
rules['lhs items'] = rules['antecedents'].apply(lambda x:len(x) )
rules[rules['lhs items']>1].sort_values('lift', ascending=False).head()
```

```python
rules['antecedents_'] = rules['antecedents'].apply(lambda a: ','.join(list(a)))
rules['consequents_'] = rules['consequents'].apply(lambda a: ','.join(list(a)))
# Transform the DataFrame of rules into a matrix using the confidence metric
pivot = rules[rules['lhs items']>1].pivot(index = 'antecedents_',
                    columns = 'consequents_', values= 'confidence')
# Generate a heatmap with annotations
sns.heatmap(pivot, annot = True)
plt.title('Heat Map - For Confidence Metric')
plt.yticks(rotation=0)
plt.xticks(rotation=90)
```

```python
rules_li = association_rules(frequent_itemsets, metric="lift", min_threshold=0.6)
print(rules_li.shape)
```

```python
rules_li['lhs items'] = rules_li['antecedents'].apply(lambda x:len(x) )
rules_li[rules_li['lhs items']>1].sort_values('lift', ascending=False).head()
```

```python
 #Replace frozen sets with strings
rules_li['antecedents_'] = rules_li['antecedents'].apply(lambda a: ','.join(list(a)))
rules_li['consequents_'] = rules_li['consequents'].apply(lambda a: ','.join(list(a)))
# Transform the DataFrame of rules into a matrix using the lift metric
pivot = rules_li[rules_li['lhs items']>1].pivot(index = 'antecedents_',
                    columns = 'consequents_', values= 'lift')
# Generate a heatmap with annotations on and the colorbar off
sns.heatmap(pivot, annot = True)
plt.title('Heat Map - For Lift Metric')
plt.yticks(rotation=0)
plt.xticks(rotation=90)

```

# apirori algorithm

```python
frequent_itemsets.ap = apriori(movie1, min_support=0.15, use_colnames=True,verbose=1)
print(frequent_itemsets.ap.head())
```

```python
frequent_itemsets.ap.sort_values("support", ascending = False).shape
```

```python
rules.ap = association_rules(frequent_itemsets.ap, metric="confidence", min_threshold=0.1)
print(rules.ap.head())
```

```python
rules.ap[(rules.ap.support > 0.015) & (rules.ap.confidence > 0.4)].sort_values("confidence", ascending = False).shape
```

```python
rules.ap['lhs items'] = rules.ap['antecedents'].apply(lambda x:len(x) )
rules.ap[rules.ap['lhs items']>1].sort_values('lift', ascending=False).head()
```

```python
rules.ap['antecedents_'] = rules.ap['antecedents'].apply(lambda a: ','.join(list(a)))
rules.ap['consequents_'] = rules.ap['consequents'].apply(lambda a: ','.join(list(a)))
# Transform the DataFrame of rules into a matrix using the confidence metric
pivot = rules.ap[rules.ap['lhs items']>1].pivot(index = 'antecedents_',
                    columns = 'consequents_', values= 'confidence')
# Generate a heatmap with annotations
sns.heatmap(pivot, annot = True)
plt.title('Heat Map - For Confidence Metric')
plt.yticks(rotation=0)
plt.xticks(rotation=90)
```

```python
rules.ap_li['lhs items'] = rules.ap_li['antecedents'].apply(lambda x:len(x) )
rules.ap_li[rules.ap_li['lhs items']>1].sort_values('lift', ascending=False).head()
```
