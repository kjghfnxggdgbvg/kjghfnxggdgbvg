```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
data = pd.read_csv('/content/wine.csv')
```

```python
data
```

```python
data.head()
```

```python
data.info()
```

```python
data.shape
```

```python
scaled_df = scale(data.iloc[:,:14])
```

```python
scaled_df
```

```python
model = PCA(n_components=3)
```

```python
pcomps = model.fit_transform(scaled_df)
pcomps
```

```python
exp_var = model.explained_variance_ratio_
exp_var
```

```python
np.cumsum(exp_var)
```

```python
plt.scatter(pcomps[:,0],pcomps[:,1], c=data['Type'])
```

```python
new_data = pd.DataFrame(pcomps[:,0:3])
```

```python
new_data
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters= 3)
```

```python
kmeans.fit(new_data)
```

```python
kmeans.labels_
```
